## PIE CHART TO SHOW THE DISTRIBUTION OF EACH COMPONENT.

```python
df = pd.DataFrame({'Value': [187,183,181,181,174,165,143]
                },
                  index=['PM2.5','NO2','SO2','CO','OZONE','PM10','NH3'])
plot = df.plot.pie(y='Value', figsize=(7, 7))

```

## BAR GRAPH TO SHOW THE SAME

```python
Value =[187,183,181,181,174,165,143]
index=['PM2.5','NO2','SO2','CO','OZONE','PM10','NH3']
df = pd.DataFrame({'Value': Value,
                   }, index=index)
ax = df.plot.bar(rot=0,legend='True')
```
